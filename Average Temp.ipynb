```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine, inspect
from functools import reduce
```

```python
avgtemp_file = "Resources\GlobalLandTemperaturesByCountry.csv\GlobalLandTemperaturesByCountry.csv"

avgtemp_df = pd.read_csv(avgtemp_file)

avgtemp_df.head()
```

```python
avgtemp_df['dt'] = pd.to_datetime(avgtemp_df['dt'])
avgtemp_df.head()
```

```python
avgtemp_df['day'] = avgtemp_df['dt'].dt.day
avgtemp_df['month'] = avgtemp_df['dt'].dt.month
avgtemp_df['year'] = avgtemp_df['dt'].dt.year

avgtemp_df.head()
```

```python
UNIQCountry = avgtemp_df.Country.unique()
len(UNIQCountry)
```

```python
UNIQYear = avgtemp_df.year.unique()
len(UNIQYear)
```

```python
UNIQCountry = np.insert(UNIQCountry, 0, 'year')
finaldf = pd.DataFrame(columns=UNIQCountry)
finaldf.year = UNIQYear[len(UNIQYear)-14:len(UNIQYear)]
finaldf = finaldf.set_index('year')
finaldf.head()
```

```python
for country in UNIQCountry:
    avgTemp = []
    for ind in range(len(UNIQYear)-14,len(UNIQYear)):
        avgtempgb = avgtemp_df.AverageTemperature[(avgtemp_df.Country == country) & (avgtemp_df.year == UNIQYear[ind] )].mean()
        avgTemp.append(avgtempgb)
    finaldf[country] = avgTemp
```
